```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the number of years (2023 - 1965)
years = 2023 - 1965

# Calculate CAGR (Compound Annual Growth Rate)
df['Growth Rate'] = ((df['2023'] / df['1965']) ** (1 / years) - 1) * 100  # Multiply by 100 to convert to percentage

# Get highest and lowest growth rate
highest_growth_rate = df[['Country Name', 'Growth Rate']].sort_values(by='Growth Rate', ascending=False).head(10)
lowest_growth_rate = df[['Country Name', 'Growth Rate']].sort_values(by='Growth Rate').head(10)

# Get highest and lowest population counts for 2023
highest_population_count = df[['Country Name', '2023']].sort_values(by='2023', ascending=False).head(10)
lowest_population_count = df[['Country Name', '2023']].sort_values(by='2023').head(10)

# Visualization for Growth Rates
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Top 10 Highest Growth Rates
axes[0].barh(highest_growth_rate['Country Name'], highest_growth_rate['Growth Rate'], color='green')
axes[0].set_title('Top 10 Countries with Highest Population Growth Rates (1965-2023)')
axes[0].set_xlabel('Growth Rate (%)')
axes[0].invert_yaxis()

# Top 10 Lowest Growth Rates
axes[1].barh(lowest_growth_rate['Country Name'], lowest_growth_rate['Growth Rate'], color='red')
axes[1].set_title('Top 10 Countries with Lowest Population Growth Rates (1965-2023)')
axes[1].set_xlabel('Growth Rate (%)')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

# Visualization for Population Counts
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Top 10 Highest Population Counts in 2023
axes[0].barh(highest_population_count['Country Name'], highest_population_count['2023'], color='blue')
axes[0].set_title('Top 10 Countries with Highest Population of senior citizens in 2023')
axes[0].set_xlabel('Population in Millions')
axes[0].invert_yaxis()

# Top 10 Lowest Population Counts in 2023
axes[1].barh(lowest_population_count['Country Name'], lowest_population_count['2023'], color='orange')
axes[1].set_title('Top 10 Countries with Lowest Population in 2023')
axes[1].set_xlabel('Population in Millions')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

# Print statements for growth rates and population count
highest_growth_rate, lowest_growth_rate, highest_population_count, lowest_population_count
```

```python
# Step 1: Get a list of common countries from the population dataset
df_population = pd.read_csv('latest_population_count.csv')
df_dependency_ratio = pd.read_csv('latest_old_age_dependency_ratio.csv')
df_healthcare_expenditure = pd.read_csv('latest_health_expenditure_per capita.csv')
df_life_expectancy = pd.read_csv('latest_life_expectancy.csv')
```

```python
common_countries = df_population[['Code', 'Country Name']]

# Filter Dependency Ratio dataset
df_dependency_ratio_filtered = df_dependency_ratio[df_dependency_ratio['Code'].isin(common_countries['Code'])]

# Filter Healthcare Expenditure dataset
df_healthcare_expenditure_filtered = df_healthcare_expenditure[df_healthcare_expenditure['Code'].isin(common_countries['Code'])]

# Filter Life Expectancy dataset
df_life_expectancy_filtered = df_life_expectancy[df_life_expectancy['Code'].isin(common_countries['Code'])]

# Step 3: Now that all datasets have the same countries, proceed with the merge operation

# Merge the datasets
df_merged = df_population.merge(df_dependency_ratio_filtered, on='Code', how='inner')\
                         .merge(df_healthcare_expenditure_filtered, on='Code', how='inner')\
                         .merge(df_life_expectancy_filtered, on='Code', how='inner')

# Step 4: Check the resulting merged dataframe
print(df_merged.head())

# Optional: Save the merged dataset to CSV for future use
df_merged.to_csv('merged_dataset.csv', index=False)
```
